```python
import pandas as pd
```

```python
movies = pd.read_csv('Movies.csv')
rating = pd.read_csv('UserRatings.csv')
```

```python
movies.head()
```

```python
rating = rating.drop(['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5'], axis = 1)
rating.head()
```

```python
movies['RATING_DATE'] = pd.to_datetime(movies['RATING_DATE'])
movies['year'] = pd.DatetimeIndex(movies['RATING_DATE']).year
movies.head()
```

Ratings

```python
movies.info()
```

```python
rating.info()
```

Merge

```python
data = pd.merge(rating, movies, left_on = 'MOVIE_ID', right_on = 'ID_MOVIE')
data.head()
```

```python
data.info()
```

```python
top_10_best_rating = data.groupby(['MOVIE_TITLE'])[['RATING']].mean().sort_values(by = "RATING", 
                                                                                ascending = False)
top_10_best_rating.head(10)
```

```python
top_10_worst_rating = data.groupby(['MOVIE_TITLE'])[['RATING']].mean().sort_values(by = "RATING", 
                                                                                ascending = True)
top_10_worst_rating.head(10)
```

```python
(data['year'] == data['MOVIE_YEAR']).value_counts()
```

```python
top_10_most_rated = data.groupby(['MOVIE_TITLE'])[['RATING']].count().sort_values(by = "RATING", 
                                                                                ascending = False)
top_10_most_rated.head(10)
```

```python
top_10_less_rated = data.groupby(['MOVIE_TITLE'])[['RATING']].count().sort_values(by = "RATING", 
                                                                                ascending = True)
top_10_less_rated.head(10)
```

```python
rep = data['year'] == data['MOVIE_YEAR']
rep = data[rep]
```

```python
#Mas votado en su año de salida
top_10_most_rated_y = rep.groupby(['MOVIE_TITLE'])[['RATING']].count().sort_values(by = "RATING", 
                                                                                ascending = False)
top_10_most_rated_y.head(10)
```

```python
#Mas votado en su año de salida
top_10_less_rated_y = rep.groupby(['MOVIE_TITLE'])[['RATING']].count().sort_values(by = "RATING", 
                                                                                ascending = True)
top_10_less_rated_y.head(10)
```
